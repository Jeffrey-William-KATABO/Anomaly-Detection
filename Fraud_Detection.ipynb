### Data Preprocessing



 Scaling the 'Amount' feature

```python
scaler = MinMaxScaler()
data['Amount'] = scaler.fit_transform(data[['Amount']])
data.head()
```

Splitting the data into training and testing

```python
normal_data = data[data['Class'] == 0]
X = normal_data.drop(columns=['Class', 'Time'])
X = X.sample(frac=1, random_state=42).reset_index(drop=True) # Shuffling the dataframe
X_train = X.iloc[:int(0.8*len(X))]
X_test = X.iloc[int(0.8*len(X)):]
print(len(X), len(X_train), len(X_test), len(X_train)+len(X_test))
```

```python
X = data.drop(columns=['Class', 'Time'])
y = data['Class']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, shuffle = True, random_state=42)

# Converting y_train and y_test to a Pandas Series for easier indexing
y_train = pd.Series(y_train)
y_tests = pd.Series(y_test)

# Select rows where y_train is 0
X_train0 = X_train[y_train == 0]
X_test0 = X_test[y_tests == 0]

print(X_train0.shape)
print(X_test0.shape)
```

## Building the Autoencoder Model Using TensorFlow

```python
### The following code section builds an autoencoder networks of 14 layers, each
### layer having respectively 29, 27, 25, 23, 21, 19, 17, 15, 13, 11, 13, 15, 17, 19, 21, 23, 25, 27, and 29 layers

input_dim = X_train0.shape[1] # Input dimension equal to the number of features: 29
input_layer = Input(shape=(input_dim, ))
nodes_number = 27
while (nodes_number>=9):
    if nodes_number == 27:
        layer = Dense(nodes_number, activation='relu')(input_layer)
    else:
        layer = Dense(nodes_number, activation='relu')(layer)
    nodes_number=nodes_number-2
nodes_number=11
while(nodes_number<=27):
    layer = Dense(nodes_number, activation='relu')(layer)
    nodes_number=nodes_number+2
output_layer = Dense(input_dim, activation='relu')(layer)
autoencoder1 = Model(inputs=input_layer, outputs=output_layer)
autoencoder1.summary()
autoencoder1.compile(optimizer='adam', loss='mean_squared_error')
```

### Training and Evaluating the Model

```python
# Defining callbacks to control the training: when the model reaches a desired level
# of performance (when the reconstruction loss drops under a defined threshold), the training stops
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('val_loss') <= 1.5):
      print("\Loss is lower than 1.5 so cancelling training!")
      self.model.stop_training = True
cb = myCallback()
```

```python
# Early stopping to prevent overtraining and to ensure decreasing validation loss
early_stop = EarlyStopping(monitor='val_loss',
                           patience=3,
                           restore_best_weights=True,
                           mode='min')
```

```python
history1=autoencoder1.fit(X_train0, X_train0,
                epochs=30,
                callbacks=[early_stop],
                batch_size=32,
                validation_data=(X_test0, X_test0))
```

```python
autoencoder1.evaluate(X_test0, X_test0)
```

```python
threshold = 0.83
def classify(X, model):
  predictions = []
  X_pred = model.predict(X)
  X = X.to_numpy()
  for i in range(len(X)):
    if tf.keras.losses.MSE(X[i], X_pred[i]).numpy() > threshold:
      predictions.append(1)
    else:
      predictions.append(0)
  return np.array(predictions)
```

```python
def get_metrics(y_obs, y_preds):
    TP, TN, FP, FN = 0,0,0,0
    for i in range(len(y_obs)):
        if y_obs[i] == 1 and y_preds[i] == 1:
            TP+=1
        elif y_obs[i] == 0 and y_preds[i] == 1:
            FP+=1
        elif y_obs[i] == 1 and y_preds[i] == 0:
            FN+=1
        elif y_obs[i] == 0 and y_preds[i] == 0:
            TN+=1

    accuracy = (TP+TN)/(TP+FP+TN+FN)
    precision = TP/(TP+FP)
    recall = TP/(TP+FN)
    f1 = 2*precision*recall/(precision+recall)
    return {"Accuracy":accuracy, "Precision":precision, "Recall":recall, "F1 Score":f1}
y_preds = classify(X_test, autoencoder1)
get_metrics(y_test.to_numpy(), y_preds)
```
